```python
import pandas as pd 
```

```python
df = pd.read_csv("ramen-ratings.csv")
df
```

```python
df.info() #shows data types and nulls 
```

```python
df_ramen = df.drop(['Top Ten', "Variety"], axis=1, inplace = True) #drops, top ten and Variety column 
df_ramen = df.dropna(subset = ['Review #','Brand', 'Style', 'Country','Stars']) #drops all null values within these columns
df_ramen
```

```python
df_ramen['Country'].replace({'United States': 'USA'}, inplace = True) #replaces any value United States as USA
```

```python
df_ramen.loc[573]
```

```python
df.loc[df['Stars']=='Unrated'] #locates the Unrated values in Stars column
```

```python
df_ramen.drop(df_ramen[df_ramen['Stars'] == 'Unrated'].index, inplace = True) #drop ramen that was unrated 
df_ramen
```

```python
df_ramen['Stars'] = df_ramen['Stars'].astype(float) #converts stars rating into float 
df_ramen.info()
```

```python
df_ramen.tail(10) #review numbers are placed in ascending order
```

```python
df_ramen.to_csv('Cleaned_Ramen.csv', index = False) #saves cleaned ramen dataset as csv 
```
